```python
corpus = """One woman named Lisa cultivated a reading habit by increasing her exposure to books. I've read more books by continually having 20-30 books on hold at the library, she said. It saves time on browsing for books. I always have new things to read with a three-week deadline."""
```

```python
import nltk
from nltk.stem import PorterStemmer # Used to do Stemming.(connecting a word to its base Stem word)
from nltk.corpus import stopwords
```

Tokenization involves converting COrpus to documents  and documents into words

to tokeninze we use sent_tokenize() from nltk library. it returns tokenized version of a sentence!

```python
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
tokenized_sentence = nltk.sent_tokenize(corpus)
print(tokenized_sentence)
```

```python
stemmer = PorterStemmer()
```

```python
stemmer.stem("gone")
```

The Tokenized Corpus may have things like \n \t, we need to clean it, for that:

```python
import re
```

Using Regular Expressions to remove \n \t

```python
cleaned_sentences =[]
for i in range(len(tokenized_sentence)):
    review = re.sub('[^a-zA-Z]',' ',tokenized_sentence[i])
    review = review.lower()
    cleaned_sentences.append(review)
```

```python
cleaned_sentences
```

```python
nltk.download('stopwords')
```

```python
stopwords.words('english') #Gives us list of stopwords
```

.word_tokenizer(sentence) converts input sentence/document into words

```python
stemmed_words=[]
lemmatized_words =[]
for i in cleaned_sentences:
    words = nltk.word_tokenize(i) #Converts Sentences into Words
    for word in words:
        if word not in set(stopwords.words('english')): ## Condition checks that teh word is nmt a stopword, if its a stopword, tehn its removed!
            stemmed_words.append(stemmer.stem(word))
            lemmatized_words.append(lemmatizer.lemmatize(word))
```

```python
stemmed_words
```

```python
lemmatized_words
```
