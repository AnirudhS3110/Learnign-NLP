```python
import gensim
from gensim.models import Word2Vec , KeyedVectors
```

### WE are gonna use Pre-Trained model word2vec-google-news-300

### This means this model gives vector of Dimension 300 as output

```python
import gensim.downloader as api
wv = api.load('word2vec-google-news-300')
```

### .most_similar(word) gives list of similar words to teh given input word, along with their Vector Embedding.

```python
wv.most_similar('king')
```

### For using cosine similarity we shd use .similarity(word1,word2), this gives how similar those two words are

wv.similarity('man','king')
